##  이 때, 원하는 데이터만 추출하여 새로운 리스트를 만들어 반환하는 함수를 작성하시오.

- 1. 응답을 json 형식으로 변환합니다.
- 2. key 값이 "result" 인 데이터를 출력합니다.
- 3. 위의 결과 중 key 값이 "optionList" 인 데이터를 변수에 저장합니다.
- 4. 3번에서 저장된 값을 반복하며, 원하는 데이터만 추출 및 가공하여 결과 리스트에 저장합니다.

```python
api_key = 'e00254010190c95c7e2a1a4fa3c2b0db'
    
url = f'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json'
params = {
    'auth' : api_key ,
    'topFinGrpNo' : '020000',
    'pageNo': 1
}
    
response = requests.get(url,params = params).json()
```

```python
# 아래 코드는 수정하지 않습니다.

if __name__ == '__main__':
    
    #  key 값이 "optionList" 인 데이터를 변수 p3 에 저장
    p3 = response['result']['optionList']
        
    # prrint.prrint(): json 을 보기 좋은 형식으로 출력
    pprint.pprint(p3)
```

```python
# 각 딕셔너리의 key를 변경(영->한)하는 함수 생성

def change_keys(p3, key_map):

    new_p3 =[]
    
    for dictionary in p3:
        new_dict = {key_map.get(key, key): value for key, value in dictionary.items()}
        new_p3.append(new_dict)
        
    return new_p3


# 키 이름 변경을 위한 매핑 딕셔너리 생성
key_map = {'fin_prdt_cd': '금융상품코드', 'intr_rate': '저축 금리', 
           'save_trm': '저축 기간', 'intr_rate_type':'저축금리유형',
           'intr_rate_type_nm':'저축금리유형명','intr_rate2':'최고 우대금리'}

# 키 이름을 변경하여 새로운 리스트 생성
new_p3 = change_keys(p3, key_map)


# 딕셔너리의 key 데이터를 일부 삭제
for i in new_p3:
    del i['dcls_month'], i['fin_co_no']

    
# pprint 이용하여 가독성있게 출력.
pprint.pprint(new_p3)
```
