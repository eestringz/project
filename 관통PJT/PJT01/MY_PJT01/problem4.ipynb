```python
# 각 딕셔너리의 key를 변경(영->한)하는 함수 생성

def change_keys(p3, key_map):

    new_p3 =[]
    
    for dictionary in p3:
        new_dict = {key_map.get(key, key): value for key, value in dictionary.items()}
        new_p3.append(new_dict)
        
    return new_p3


# 키 이름 변경을 위한 매핑 딕셔너리 생성
key_map = {'fin_prdt_cd': '금융상품코드', 'intr_rate': '저축 금리', 
           'save_trm': '저축 기간', 'intr_rate_type':'저축금리유형',
           'intr_rate_type_nm':'저축금리유형명','intr_rate2':'최고 우대금리'}

# 키 이름을 변경하여 새로운 리스트 생성
new_p3 = change_keys(p3, key_map)


# 딕셔너리의 key 데이터를 일부 삭제
for i in new_p3:
    del i['dcls_month'], i['fin_co_no']

    
# pprint 이용하여 가독성있게 출력.
pprint.pprint(new_p3)
```
